    One of the most crucial aspects of data analysis is dealing with missing data. Missing values can cause bias, weaken statistical power and affect the reliability of results. In this article, we will delve into the significance, techniques, challenges and implications of handling missing data in a real-world data analysis scenario.

#### Significance of Handling Missing Data:
    1.) Biased Results: Ignoring missing data can lead to biased results, particularly if the missing data is not missing completely at random (MCAR).
    
    2.) Reduced Statistical Power: Reduced statistical power due to missing data can make it challenging to identify authentic effects during analyses.
    
    3.) Inaccurate Inferences: Inaccurate inferences about relationships within data can result from mishandling of missing data.

#### Techniques for Handling Missing Data:
    1.) Deletion Methods: If there are only a few missing values and they are distributed at random, it is possible to remove rows or columns using the dropna() function in Pandas.
    
    2.) Imputation: Replacing empty values with calculated or estimated values is a common practice. The techniques used for this include filling in with mean, median, or mode imputation, or more advanced strategies such as regression imputation.

#### Challenges Posed by Missing Data:
    1.) Data Bias: It is possible that the reason for the absence of data is connected with the variable, which can result in bias. For instance, individuals who earn a high income and opt not to disclose it may have missing income data.
    
    2.) Increased Variability: It is possible that the imputed data's variability is not fully accounted for, which could result in standard errors and confidence intervals that are overly optimistic.

#### Potential Implications of Different Imputation Methods:
    1.) Mean Imputation:
          Pros: Simple and quick.
          Cons: If the missing data are not MCAR, this method may not be appropriate. In addition, if missing values are systematically associated with other variables, it has the potential to distort relationships.

    2.) Regression Imputation:
          Pros: This approach takes into account the relationships with other variables.
          Cons: It is assumed that there is a linear relationship, but this assumption may create bias if there is actually a nonlinear relationship.

    3.) Multiple Imputation:
          Pros: Imputation uncertainty is considered in the accounts.
          Cons: It may require significant computational resources to process this task.

#### Importance of Assessing and Documenting Handling of Missing Data:
    1.) Transparency: Providing documentation on how missing data is handled can improve the transparency of the analysis, enabling others to comprehend and reproduce the procedure.
    
    2.) Reproducibility: Reproducibility is ensured by maintaining a transparent log of missing data handling, which permits others to validate and expand on the analysis without any issues.

```python
# EXAMPLES

import pandas as pd
import numpy as np

# Create a DataFrame with missing values
data = {'A': [1, 2, np.nan, 4], 'B': [5, np.nan, 7, 8]}
df = pd.DataFrame(data)

# Scenario 1: Delete Rows with Missing Values
df_deleted_rows = df.dropna()

# Scenario 2: Impute with Mean
df_imputed_mean = df.fillna(df.mean())

# Scenario 3: Impute with Regression
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression

imputer = SimpleImputer(strategy='mean')
reg_imputer = SimpleImputer(strategy='mean')

df_imputed_reg = pd.DataFrame(reg_imputer.fit_transform(df), columns=df.columns)

print("Original DataFrame:")
print(df)

print("\nDataFrame after deleting rows with missing values:")
print(df_deleted_rows)

print("\nDataFrame after imputing with mean:")
print(df_imputed_mean)

print("\nDataFrame after regression imputation:")
print(df_imputed_reg)
```

# Data Cleaning

```python
import pandas as pd
import numpy as np
df=pd.read_csv('Cricket Dataset.csv')
df = df.dropna(how= 'all', axis = 0 )# Deleting totally blank rows
df = df.dropna(how= 'all', axis = 1 )# Deleting totally blank columns
df.rename(columns=df.iloc[0], inplace = True) #Renaming the column names with the 1st row values
df.drop(df.index[0], inplace = True) #Dropping the row which is made as column names
df.columns = df.columns.str.strip() # stripping spaces from column names and reassigning the clean column names to the dataset
df=df.replace(' ', np.NaN) # replacing any spaces with Null values so that we can use fillna method
df=df.replace('', np.NaN)
df = df.dropna(subset = ['Match Number','Number of Wickets Taken']) # Deleting all rows where Match Number and Number of Wickets Taken are empty. 
#df = df.fillna(method = 'ffill')  #Forward filling the dataset
df=df.ffill()

# Splitting of Player Name column into thier First Name, Title and Last Name respectively

df['First Name'] = df['Player Name'].str.split().str[0]
df['Title'] = df['Player Name'].loc[df['Player Name'].str.split().str.len() == 3].str.split(expand=True)[1]
df['Last Name'] = df['Player Name'].str.split().str[-1]
df['Title'] = df['Title'].replace(np.nan,'(NA)') # Formatting null values as (NA) as per the required output file

# Splitting Salary column with '-' as delimiter and assigning first and second part as minimum and maximum salary respectively

df['Minimum_Salary']=df['Salary Estimate'].str.split(pat='-').str[0] 
df['Maximum_Salary']=df['Salary Estimate'].str.split(pat='-').str[1]

# passing the min and max salary columns to df.replace(), specifying each char and it's replacement to obtain the integer values alone

cols = ['Minimum_Salary', 'Maximum_Salary']
df[cols] = df[cols].replace({'\$': '','K': '','\([A-Z][a-z]+\s\w+\.\)':''}, regex=True)

# Finding average of the salary

df['Minimum_Salary'] = df['Minimum_Salary'].astype(int)
df['Maximum_Salary'] = df['Maximum_Salary'].astype(int)
df['Average_Salary']=(df['Minimum_Salary']+ df['Maximum_Salary'])/2
df['Average_Salary'] = np.ceil(df['Average_Salary']).astype(int) #Ceiling function maps x value to the least integer greater than or equal x

# To export the output to a csv file

df.to_csv('Cleaned_file.csv')
df
```

## Left Join

    When merging two datasets, a left join includes all the rows from the left dataframe as well as the matching rows from the right dataframe depending on the provided key(s). If there is no match in the appropriate dataframe, the columns in the resultant dataframe will contain NaN values. In the below example left join gets us the performance detail of each player who has been assigned/played any of the matches along with their profile info like their age, team, role etc.,

```python
left_merged=pd.merge(df,player_info_df,on=['Player Code','Player Name'],how='left')
left_merged
```

## Right Join

    1. The primary distinction is whether all rows of a dataframe are included in the result. It is the right dataframe in a right join and the left dataframe in a left join. The choice between right and left join is determined by which dataframe's data you wish to keep entirely in the merged result.

    2. Whenever we need to find the players who are not assigned/played any matches but still a part of the team as coaches/substitutes, we can use right join to get the all the rows from player info table with/without their match information from our left table.

```python
right_merged=pd.merge(df,player_info_df,on=['Player Code','Player Name'],how='right')
right_merged
```

## Outer Join

    1. An outside join is often used in data merging when you wish to incorporate all rows from both datasets, regardless of whether there is a match based on the provided key(s). It is useful in situations where you want to build a merged dataset that keeps information from both dataframes, filling in missing values with NaN for dataframe columns where no match is found.

    2. We can use outer join to find the average of batting strike rate or bowling economy rate average or average salary of overall team members. For example,

```python
outer_merged=pd.merge(df,player_info_df,on=['Player Code','Player Name'],how='outer')
outer_merged
```

## Inner Join

    1. An inner merge information on these data sets can help us to identify each player's age, which provides insight into the age distribution of players in relation to their performance and salary. We may also examine how different player positions (batsman, bowler, all-rounder), batting strike rates, and bowling economy rates match, as well as providing insights into their performance metrics, pay, and other facets of the game.

    2. 
    a.) The Difficulty of Analyzing Outliers:  Outliers or exceptional cases in one table that are not present in the other will be excluded from the analysis. This could be a problem if you're looking for outliers or unusual cases.
    b.) Identifying Data Quality Issues is Difficult:  An inner join may obscure concerns with data quality, such as missing or inconsistent key values, making it more difficult to discover and address these issues in the combined dataset.
    c.) Data loss: The most major disadvantage of an inner join is that it rejects rows that do not have matching keys in both tables. Players who appear in either Table 1 or Table 2 but not both will be eliminated from the results. This could result in the loss of crucial information, especially if you want to analyze all participants, regardless of whether they have complete information in both tables.

```python
inner_merged=pd.merge(df,player_info_df,on=['Player Code','Player Name'],how='inner')
inner_merged
```
